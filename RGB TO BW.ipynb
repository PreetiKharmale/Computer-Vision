```python
import cv2
#     https://www.geeksforgeeks.org/image-processing-in-python/?ref=lbp         ----- input image
```

```python
#read image
img = cv2.imread('/content/Ganeshji.webp', cv2.IMREAD_UNCHANGED)                                             # UNCHNAGE - ALPHA VALUE MUSTNOT BE CHANGED

# get dimension of the img
dimensions = img.shape

# height, width, channels
height = img.shape[0]
width = img.shape[1]
channels = img.shape[2]
print('Image Dimension : ' , dimensions)
print('Image Height : ' , height)
print('Image Width : ' , width)
print('Numbernof Channels  : ' , channels)

```

```python
# Splitting the image into red, green , blue
```

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/Ganeshji.webp',1)
b,g,r = cv2.split(img)

cv2_imshow(img)
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
```
