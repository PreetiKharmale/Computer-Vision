## **Erosion**
Removes pixels on the boundaries of objects in an image, which will make the objects smaller.

```python
# we are coverting color img to gray scale , coz color img has many channels
```

```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/download.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))                                 # morph rectangle

eroded1 = cv2.erode(gray, kernel, iterations=1)
eroded2 = cv2.erode(gray, kernel, iterations=2)

cv2_imshow(image)
cv2_imshow(eroded1)
cv2_imshow(eroded2)
cv2.waitKey(0)
```

# **Dilation**.
It adds pixels on the boundaries of objects in an image, which will increase the size of objects in the image.

```python
image = cv2.imread("/content/download.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

dilate1 = cv2.dilate(gray, kernel, iterations=1)
dilate2 = cv2.dilate(gray, kernel, iterations=2)

cv2_imshow(image)
cv2_imshow( dilate1)
cv2_imshow(dilate2)
cv2.waitKey(0)
```

#**Opening**
A morphological opening is an erosion followed by a dilation. This operation is used to remove small objects and noise from an image.

```python
opening1 = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel, iterations=1)
opening2 = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel, iterations=2)

cv2_imshow( image)
cv2_imshow(opening1)
cv2_imshow(opening2)
cv2.waitKey(0)
```

# **Closing**
A morphological closing is a dilation followed by an erosion. This operation is used to close small holes inside the objects.

```python
closing1 = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel, iterations=1)
closing2 = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel, iterations=2)

cv2_imshow(image)
cv2_imshow(closing1)
cv2_imshow(closing2)
cv2.waitKey(0)
```
